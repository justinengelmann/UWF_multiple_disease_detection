```python
expname = 'r34_REp03s3pix' 

n_epochs = 40

labelsmoothing_alpha = 0.05
weight_decay = 5e-6
dropouts = [0.1,0.5]
aug_prob = 0.9
use_cosine_scheduler=True
use_timm_scheduler=False
lr = 5e-5 if not use_cosine_scheduler else 0.1
T_0 = 10
T_mult = 1
use_ema=True
use_mixup=True

expname += f"{'aug_' if aug_prob>0. else ''}{'drop_' if any((d>0 for d in dropouts)) else ''}"
expname += f"L2{str(weight_decay)[-1]}_{'aLS_' if labelsmoothing_alpha>0 else''}"
if use_ema: expname+='ema_'
if use_mixup: expname+='mixupp1a06_'
expname += f"{'cos'+str(T_0)+str(T_mult) if labelsmoothing_alpha>0 else''}{'timm' if use_timm_scheduler else ''}_{str(n_epochs)+'eps'}"

include_third_channel = False
batch_size = 32

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
import timm 
from timm.scheduler.cosine_lr import CosineLRScheduler
from timm.utils.model_ema import ModelEmaV2
# from timm.data.mixup import Mixup
# import patched Mixup which supports multi-label case
from mixup_patched import Mixup
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import StratifiedKFold
from datetime import datetime
from tqdm.notebook import tqdm
from pathlib import Path

seed=42
np.random.seed(seed)
torch.manual_seed(seed)
expname+=datetime.today().strftime('_%m_%d')
expname=Path(f'expresults/{expname}')
expname.mkdir(exist_ok=True)

class TOPDataset(Dataset):
  def __init__(self, csv_file, img_dir, targets, transform=None,
               covariates=None, return_covariates='auto', 
               labelsmoothing_alpha=0, include_third_channel=False):
    self.csv_file = csv_file
    self.img_dir = Path(img_dir)
    self.transforms = transforms.ToTensor() if transform is None else transform
    self.targets = targets if not isinstance(targets, str) else [targets]
    self.covariates = covariates if not isinstance(covariates, str) else [covariates]
    self.return_covariates = (covariates is not None) if return_covariates=='auto' else return_covariates
    self.labelsmoothing_alpha = labelsmoothing_alpha
    self.include_third_channel = include_third_channel
    self.df = pd.read_csv(csv_file)
    
  def __len__(self):
    return self.df.filename.size
  
  def __getitem__(self, idx):
    row = self.df.iloc[idx,:]
    
    image = Image.open(self.img_dir/row['filename'])
    # all images are loaded as left eyes by default, i.e. right eyes are flipped
    if row['LR'] == 'R':
      image=image.transpose(Image.FLIP_LEFT_RIGHT)
    image = self.transforms(image)
    if not self.include_third_channel:
      image=image[:2,...]
      
    targets = torch.Tensor(row[self.targets])
    
    if self.labelsmoothing_alpha > 0:
      # asymmetric smoothing
      targets *= (1-(self.labelsmoothing_alpha+0.01))
      targets += self.labelsmoothing_alpha
    
    if self.return_covariates:
      if self.covariates is not None:
        covariates = row[self.covariates].values
      else:
        covariates = None
      return image, targets, covariates
    
    return image, targets

# TOP Train set, exact
## Adding a third value for compatability with plotting images
norm_means = [0.22578795, 0.23797078, 1]
norm_stds = [0.14651306, 0.11282759, 1]


basic_transform = transforms.Compose([transforms.ToTensor(),
                                      transforms.Normalize(norm_means, norm_stds)])

# for visualisation purposes
def add_dummy_channel_if_needed(X):
  if X.shape[0]==3:
    return X
  dummy_channel = torch.zeros_like(X[0]).unsqueeze(0)
  dummy_channel -= norm_means[2]/norm_stds[2]
  return torch.cat((X,dummy_channel))
denormalize = transforms.Compose([transforms.Lambda(add_dummy_channel_if_needed),
  transforms.Normalize(-np.array(norm_means)/np.array(norm_stds),
                                                       1/np.array(norm_stds))])

def random_scale(X):
  # uniform([0.75, 1.25])
  # scale_range = 0.25
  # scale_range = 1/(scale_range*2)
  # > 2.0 is the factor we want to achieve +-0.25
  # scale = (torch.rand(3,1,1)/scale_range)+1-(1/scale_range/2)
  scale = (torch.rand(3,1,1)/2)+1-(1/2/2)
  return X * scale

def add_gaussian_noise(X):
  # Normal(0, 0.1)
  noise = torch.randn(X.size()) * 0.1
  return X + noise

augment_transforms = transforms.Compose([
      transforms.RandomErasing(p=0.3, scale=(0.05, 0.3), value='random'),
      transforms.RandomAffine(degrees=15, scale=[0.8, 1.2], shear=10),
])
domain_transforms = transforms.Compose([
      transforms.GaussianBlur(kernel_size=(7), sigma=(0.1,1)),
      transforms.Lambda(random_scale),
      transforms.Lambda(add_gaussian_noise),
])
flip_transforms = transforms.Compose([transforms.RandomHorizontalFlip(p=0.3),
                                      transforms.RandomVerticalFlip(p=0.1)])
train_transforms=transforms.Compose([
      transforms.ToTensor(),
      transforms.RandomApply([augment_transforms], p=aug_prob),
      transforms.Normalize(norm_means, norm_stds),
      flip_transforms,
      transforms.RandomApply([domain_transforms], p=aug_prob)
])
if use_mixup:
  mixup = Mixup(mixup_alpha=0.6, prob=1, label_smoothing=0)

targets = ['MH', 'RP', 'AMD', 'RVO', 'RD', 'Gla', 'DR', 'any_retina_disease']

train_set = TOPDataset(csv_file='PATH/TO/PREPROCESSED/DATA/df_train.csv',
                       img_dir = 'PATH/TO/PREPROCESSED/DATA/images',
                       targets=targets, transform=train_transforms,
                       labelsmoothing_alpha=labelsmoothing_alpha, include_third_channel=include_third_channel)
val_set = TOPDataset(csv_file='PATH/TO/PREPROCESSED/DATA/df_val.csv',
                       img_dir = 'PATH/TO/PREPROCESSED/DATA/images',
                       targets=targets, transform=basic_transform, include_third_channel=include_third_channel)
train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True, num_workers=7)
val_loader = DataLoader(val_set, batch_size=batch_size, shuffle=False, num_workers=7)


model = timm.create_model('resnet34', pretrained=True, in_chans=2)
model.fc = nn.Sequential(nn.Dropout(p=dropouts[0]), nn.Linear(512, 128), nn.PReLU(128), 
                         nn.Dropout(p=dropouts[1]), nn.Linear(128, len(targets)))

# model.fc = nn.Sequential(nn.Dropout(p=dropouts[0]), nn.Linear(512, len(targets)))

param_count = sum(p.numel() for p in model.parameters() if p.requires_grad)

criterion = nn.BCEWithLogitsLoss(reduction='none')
optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)#
scheduler = None
if use_cosine_scheduler:
  print(f'Using Cosine Schedule lr={lr}, T_0={T_0}, T_mult={T_mult}')
  optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9, weight_decay=weight_decay)
  if not use_timm_scheduler:
    scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0, T_mult)
  else:
    scheduler = CosineLRScheduler() # TODO

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model.to(device)
if use_ema:
  ema = ModelEmaV2(model=model, decay=0.999)
print(f'Training on {device}. Model has {param_count} parameters.\n'
      f'Experiment is called {expname}.')



train_losses_iter = []
val_losses_iter = []
train_aucs = np.empty(shape=(0, len(targets)))
val_aucs = np.empty(shape=(0, len(targets)))
train_losses = np.empty(shape=(0, len(targets)))
val_losses = np.empty(shape=(0, len(targets)))
n_iters_per_epoch = len(train_loader)

for epoch in range(n_epochs):
  # training
  model.train()
  running_losses = torch.zeros(len(targets), device=device)
  train_targets = torch.tensor([])
  train_preds = torch.tensor([])
  for batch_idx, (inputs, labels) in enumerate(tqdm(train_loader, desc=f'Train Ep {epoch+1:4}')):
    inputs = inputs.to(device)
    labels = labels.to(device)
    # timm's mixup only works on even-sized batches,
    # but final epoch batch might be uneven
    if use_mixup and inputs.shape[0]%2==0:
      inputs, labels = mixup(inputs, labels)
    optimizer.zero_grad()
    outputs = model(inputs)
    loss = criterion(outputs, labels)
    loss.mean().backward()
    optimizer.step()
    if use_ema:
      ema.update(model)
    if scheduler is not None:
      scheduler.step(epoch + batch_idx / n_iters_per_epoch)
    
    running_losses += loss.sum(axis=0).detach()
    train_losses_iter.append(loss.mean().detach().item())
    train_targets = torch.cat((train_targets, labels.to('cpu').round()), dim=0)
    train_preds = torch.cat((train_preds, outputs.detach().to('cpu')), dim=0)
    
  epoch_losses = running_losses.to('cpu') / len(train_set)
  train_losses = np.vstack([train_losses, epoch_losses])
  for t, l in zip(targets, epoch_losses):
    print(f'{t[:3]:3}: {l:.4f}', end=' ')
  print()
  aucs = [roc_auc_score(train_targets[:,i], train_preds[:,i]) for i in range(len(targets))]
  train_aucs = np.vstack([train_aucs, aucs])
  for t, l in zip(targets, aucs):
    print(f'{t[:3]:3}: {l:.4f}', end=' ')
    
  # validation
  if use_ema:
    eval_model = ema.module
  else:
    eval_model=model
  eval_model.eval()
  running_losses = torch.zeros(len(targets), device=device)
  val_targets = torch.tensor([])
  val_preds = torch.tensor([])
  for inputs, labels in tqdm(val_loader, desc=f'Val Ep {epoch+1:4}'):
    with torch.no_grad():
      inputs = inputs.to(device)
      labels = labels.to(device)
      outputs = eval_model(inputs)
      loss = criterion(outputs, labels)
      running_losses += loss.sum(axis=0)
      val_losses_iter.append(loss.mean().item())
      val_targets = torch.cat((val_targets.round(), labels.to('cpu')), dim=0)
      val_preds = torch.cat((val_preds, outputs.to('cpu')), dim=0)
  epoch_losses = running_losses.to('cpu') / len(val_set)
  val_losses = np.vstack([val_losses, epoch_losses])
  for t, l in zip(targets, epoch_losses):
    print(f'{t[:3]:3}: {l:.4f}', end=' ')
  print()
  aucs = [roc_auc_score(val_targets[:,i], val_preds[:,i]) for i in range(len(targets))]
  val_aucs = np.vstack([val_aucs, aucs])
  for t, l in zip(targets, aucs):
    print(f'{t[:3]:3}: {l:.4f}', end=' ')
  
  # plotting
  if all([(epoch+1)%10==0]) or epoch==(n_epochs-1):
    plt.figure(figsize=(5,1.5), dpi=120)
    l = plt.plot(np.arange(1, epoch+2), val_losses.mean(1))
    plt.plot(np.arange(1, epoch+2), train_losses.mean(1), c=l[0].get_color(), linestyle=':')
    plt.ylim(top=0.4, bottom=0)
    plt.xlim(0.5, n_epochs+0.5)
    plt.grid(True)
    plt.show()
    
    plt.figure(figsize=(5,1.5), dpi=120)
    l = plt.plot(np.arange(1, epoch+2), val_aucs[:,-1])
    plt.plot(np.arange(1, epoch+2), train_aucs[:,-1], c=l[0].get_color(), linestyle=':')
    plt.axhline(0.930590, c='gray', linestyle=':')
    plt.ylim(top=1, bottom=0.85)
    plt.xlim(0.5, n_epochs+0.5)
    plt.grid(True)
    plt.show()
    
#     plt.figure(figsize=(5,1), dpi=100)
#     plt.ylim(top=0.25, bottom=0.)
#     plt.xlim(left=0., right=len(train_losses_iter)+1)
#     train_losses_iter_series = pd.Series(train_losses_iter)
#     train_losses_iter_std = train_losses_iter_series.rolling(30).std()
#     train_losses_iter_mean = train_losses_iter_series.rolling(30).mean()
#     plt.plot(train_losses_iter_mean, c='red')
#     plt.fill_between(np.arange(len(train_losses_iter_series)),
#                      train_losses_iter_mean-train_losses_iter_std,
#                      train_losses_iter_mean+train_losses_iter_std, 
#                      alpha=0.1, color='red')
#     plt.grid(True)
#     plt.show()
    
    
plt.figure(figsize=(12,3), dpi=150)
for i, t in enumerate(targets):
  l = plt.plot(train_aucs[:, i], linestyle=':')
  plt.plot(val_aucs[:, i], label=t[:3], c=l[0].get_color())
plt.ylim(top=1, bottom=0.75)
plt.xlim(-0.5, n_epochs-0.5)
plt.legend(loc=(1,-0.025), fontsize='x-large')
plt.grid(True)
plt.show()

plt.figure(figsize=(12,3), dpi=150)
for i, t in enumerate(targets):
  l = plt.plot(train_losses[:, i], linestyle=':')
  plt.plot(val_losses[:, i], label=t[:3], c=l[0].get_color())
plt.ylim(top=0.6, bottom=0.)
plt.xlim(-0.5, n_epochs-0.5)
plt.legend(loc=(1,-0.025), fontsize='x-large')
plt.grid(True)
plt.show()


# save results
import joblib
results_dict = {
  'train_aucs': train_aucs,
  'val_aucs': val_aucs,
  'train_losses': train_losses,
  'val_losses': val_losses
}
joblib.dump(results_dict, expname/'resultsdict.pkl')
torch.save(model, expname/'model.pt')
np.save(expname/'train_aucs', train_aucs)
np.save(expname/'val_aucs', val_aucs)
np.save(expname/'train_losses', train_losses)
np.save(expname/'val_losses', val_losses)


print(expname)
print('Done\n\n')

# Save EMA model, reprint normal model val metrics, eval EMA model
if use_ema:
  torch.save(ema.module, expname/'ema_model.pt')

  for t, l in zip(targets, epoch_losses):
    print(f'{t[:3]:3}: {l:.4f}', end=' ')
  print()
  for t, l in zip(targets, aucs):
    print(f'{t[:3]:3}: {l:.4f}', end=' ')


  model.eval()
  running_losses = torch.zeros(len(targets), device=device)
  val_targets = torch.tensor([])
  val_preds = torch.tensor([])
  for inputs, labels in tqdm(val_loader, desc=f'EMA Val'):
    with torch.no_grad():
      inputs = inputs.to(device)
      labels = labels.to(device)
      outputs = model(inputs)
      loss = criterion(outputs, labels)
      running_losses += loss.sum(axis=0)
      val_losses_iter.append(loss.mean().item())
      val_targets = torch.cat((val_targets.round(), labels.to('cpu')), dim=0)
      val_preds = torch.cat((val_preds, outputs.to('cpu')), dim=0)
  epoch_losses = running_losses.to('cpu') / len(val_set)
  val_losses = np.vstack([val_losses, epoch_losses])
  for t, l in zip(targets, epoch_losses):
    print(f'{t[:3]:3}: {l:.4f}', end=' ')
  print()
  aucs = [roc_auc_score(val_targets[:,i], val_preds[:,i]) for i in range(len(targets))]
  val_aucs = np.vstack([val_aucs, aucs])
  for t, l in zip(targets, aucs):
    print(f'{t[:3]:3}: {l:.4f}', end=' ')
```
